```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

## Since our main task have to do with only Hispanic and White non-Hispanic customers, we will remove rows that are other ethnicity except the ones we are analyzing.

```python
races = ['White not Hispanic', 'Hispanic']
df = df[df['Ethnicity'].isin(races)]
```

## Remove rows that have age has a negative integer

```python
df = df[df['Age'] > 0]
```

## Remove missing values

```python
df = df.dropna()
```

## Age Cohort

```python
# New column
df['Age Cohort'] = df['Age']

def age_group(age):
    if age <= 5:
        return '0-5'
    elif age <= 12:
        return '6-12'
    elif age <= 17:
        return '13-17'
    elif age <= 21:
        return '18-21'
    elif age <= 50:
        return '22-50'
    else:
        return '51+'
df['Age Cohort'] = df['Age'].apply(age_group)
```

# Remove outliers in expenditures using IQR

```python
Q1 = df['Expenditures'].quantile(0.25)
Q3 = df['Expenditures'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

rows_before = len(df)

df = df[(df['Expenditures'] >= lower) & (df['Expenditures'] <= upper)]

rows_after = len(df)

outliers = rows_before - rows_after

print("Outliers removed:", outliers)
```

# Now we have prepared the data. Now we analyze it and find some patterns.

## Average expenditure for all males

```python
mean = df[df['Gender'] == 'Male']['Expenditures'].mean()

print("Average expenditure for all male:", mean.round(2))
```

## Average expenditure for Hispanics

```python
mean_h = df[df['Ethnicity'] == 'Hispanic']['Expenditures'].mean()

print('Average expenditure for Hispanics:',mean_h.round(2))
```

## Average expenditure for White non-Hispanic males

```python
mean_w = df[(df['Ethnicity'] == 'White not Hispanic') & (df['Gender'] == 'Male')]['Expenditures'].mean()

print('Average expenditure for White non-Hispanic males:', mean_w.round(2))
```

## Median expenditure for each age cohort

```python
df.groupby("Age Cohort")['Expenditures'].median()
```

## Which gender has highest average expenditure

```python
df.groupby('Gender')['Expenditures'].mean().round(2)
```

## Difference between genders

```python
male_avg = df[df["Gender"] == "Male"]["Expenditures"].mean()
female_avg = df[df["Gender"] == "Female"]["Expenditures"].mean()

difference = abs(male_avg - female_avg)
print(difference.round(2))
```

## Total expenditures for each ethnic group

```python
df.groupby("Ethnicity")["Expenditures"].sum()
```

## Ethnic group with most consumers in age cohort 22–50

```python
df[df["Age Cohort"] == "22-50"]["Ethnicity"].value_counts()
```

## Most common age cohort for each gender

```python
df.groupby("Gender")["Age Cohort"].value_counts()
```

## Percentage of each ethnic group

```python
df["Ethnicity"].value_counts(normalize=True) * 100
```

## Compare expenditures by ethnicity

```python
df.groupby("Ethnicity")["Expenditures"].mean()
```

## Compare expenditures by ethnicity and age cohort

```python
df.groupby(["Ethnicity", "Age Cohort"])["Expenditures"].mean()
```
